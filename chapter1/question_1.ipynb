```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#10 seconds of data, 1 ms resolution
bins=10*1000
bernoulli_coinflips=np.random.uniform(0,1,size=bins)
#rate is 100Hz, which means that it's 
rate=100*0.001
print(rate)
spikes = (bernoulli_coinflips < rate).astype(int)
plt.plot(spikes[:1000])
spikes
```

```python
spikes.sum()
```

```python
def sample_bernoulli(bins=10*1000, rate=0.1):
    bernoulli_coinflips=np.random.uniform(0,1,size=bins)
    spikes=(bernoulli_coinflips < rate).astype(int)
    return spikes

trials=[sample_bernoulli() for s in range(100)]
trials

estimator=[np.mean(trial) for trial in trials]

plt.hist(estimator, bins=50)

print('Variance of the estimator is: ', np.var(estimator))
print('Theoretical variance of the estimator is: ', (0.1*0.9)/10000)
```

```python
quadratic_risk=np.mean((np.array(estimator)-0.1)**2)
quadratic_risk
```
